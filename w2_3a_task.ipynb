a)plot a bar chart for the total prices of each paid app in each category

```python
import pandas as pd
import matplotlib.pyplot as plt

ps = pd.read_csv('cleaned_playstore.csv')
paidapps = ps[ps.Type == "Paid"]
paidapps
```

```python
paidapps.Category.value_counts().size
```

```python
pa = paidapps.groupby('Category').sum()['Price'].sort_values(ascending=False)
pa
```

```python
plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
plt.bar(pa.index, pa.values)
plt.show()
```
