```python
import pandas as pd
df = pd.read_csv("googleplaystore.csv")
df.head()
```

```python
part_len=len(df)//4
part_len
```

```python
part1=df[0:part_len]
len(part1)
```

```python
part2=df[part_len:part_len*2]
len(part2)
```

```python
part3=df[part_len*2:part_len*3]
len(part3)
```

```python
part4=df[part_len*3:]
len(part2)
```

```python
import pickle
pickle.dump(part1,open("part1.csv", "wb"))
pickle.dump(part2,open("part2.csv", "wb"))
pickle.dump(part3,open("part3.csv", "wb"))
pickle.dump(part4,open("part4.csv", "wb"))
```

#slinding window

```python
dataset =pd.read_csv("googleplaystore.csv")
```

```python
train = []
period_train = 5
period_test = 2
i = 0
u = period_train
r = period_train
k = period_train + period_test
while k < len(dataset)+1:
    X = dataset.iloc[i:u, 0:]
    X1 = dataset.iloc[r:k, 0:]
    train.append([X,X1])
    if k== len(dataset):
      break
    i += period_test
    u += period_test
    k += period_test
    r += period_test
    if k > len(dataset):
        k = len(dataset)
```
